```python
import cplex
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
cov_matrix = pd.read_csv('43StocksCovariance.csv').to_numpy()[:,1:]
sp_49 = pd.read_csv('43best.csv', index_col=0)
size = sp_49.shape[0]
print(sp_49.shape)
print(cov_matrix.shape)
print(cov_matrix)
```

```python
def setproblemdata(p):
    # minimize risk while keeping the return constant
    p.objective.set_sense(p.objective.sense.minimize)  
    
    # add 49 variables: the weights for 49 stocks
    # upper boundaries are 1, lower boundaries are 0
    cnt = sp_49.shape[0]
    names = ['w'+str(i) for i in range(cnt)]
    
    p.variables.add(obj = [0]*cnt, ub = [1]*cnt, lb = [0]*cnt, names = names) 
                                
    
    my_senses = 'EG'
    my_rhs = [1,0.1]
    rows = [[names, [1]*cnt],                        # The sum of all weights is 1
           [names, list(sp_49["return"][:cnt])]]     # The expected return is larger than 5%
    print(rows)
    p.linear_constraints.add(lin_expr = rows, senses = my_senses, rhs = my_rhs)
  
    # add quadractic part in the objective (derivative of the formula related to risk)
    qmat = []
    for j in range(cnt):
        qmat.append([[i for i in range(cnt)], [2*cov_matrix[j, m] for m in range(cnt)]]) 
    
    print(qmat)
    p.objective.set_quadratic(qmat)

p = cplex.Cplex()
setproblemdata(p)
p.solve()
```

```python
p = cplex.Cplex()
setproblemdata(p)
p.solve()
```

```python
portfolio_names = []
portfolio_weights = []
for i in range(size):
    #proportion = float('%.2f'%(p.solution.get_values()[i]))

    proportion = np.round(p.solution.get_values()[i], 3)  # keep results with 3 decimals and keep weights larger than 0 
    if proportion != 0:
        portfolio_names.append(sp_49.index.values[i])
        portfolio_weights.append(proportion)

df = pd.DataFrame(columns = portfolio_names)
df.loc['Weight'] = portfolio_weights
print('Selected stocks and their weights:')
print(df)
print('\nPortfolio risk:')
print(round(p.solution.get_objective_value(),7))
```
